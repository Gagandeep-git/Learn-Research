```python
# Top down approach
# take all features and perform following steps
## 1. calculate vif
# 2. if any vif is greater than 5, eliminate the variable and update the feature list, else goto 4 with final feature list
# 3. Repeat 1 and 2
# 4. further eliminate features based on significance level (p-value)
# 5. Choose different combinations of features to increase R squared value and F statistic and reduce BIC
```

```python
# to perform steps 1 to 3, check_vif function is created

# check_vif is a function that does the following
# 1. calculate vif
# 2. if any vif is greater than 5 and maxtimes(function parameter) is greater than zero, eliminate the variable and 
# update the feature list. imaxtimes is equal to zero, print just calculated vifs and goto 4
# 3. Repeat 1 and 2
# 4. the number of times loop inside the function runs is specified by the user through parameter: maxtimes

#col = X_train.columns
X_train_for_rfe=X_train
def check_vif(X_train, maxtimes):
    #y_train=df_train.pop('price')
    #X_train=df_train
    col=X_train.columns
    column_list=list(col)
    vif_rfe = pd.DataFrame()
    X_train=X_train[column_list]
    vif_rfe['Features'] = X_train.columns

    vif_rfe['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range (X_train.shape[1])]
    vif_rfe['VIF'] = round(vif_rfe['VIF'],2)
    vif_rfe = vif_rfe.sort_values( by = 'VIF', ascending = False)
    #vif_rfe
    vif_rfe = vif_rfe.replace([np.inf, -np.inf], np.nan)
    
    
    multicollinearity_detected = False
    #
    if vif_rfe.iloc[0].VIF>=5 | np.isnan(vif_rfe.iloc[0].VIF):
        #print('inside if')
        multicollinearity_detected=True
    
    #maxtimes=1
    #print(maxtimes)
    while(maxtimes>0 & multicollinearity_detected):
        #print('inside while')
        column_list.remove(vif_rfe.iloc[0].Features)
        X_train_new=pd.DataFrame()
        X_train_new=X_train[column_list]
        X_train=pd.DataFrame()
        X_train=X_train_new
        
        vif_rfe = pd.DataFrame()
        #X_train_new=X_train[col_list]
        vif_rfe['Features'] = X_train.columns

        vif_rfe['VIF'] = [variance_inflation_factor(X_train_new.values, i) for i in range (X_train_new.shape[1])]
        vif_rfe['VIF'] = round(vif_rfe['VIF'],2)
        vif_rfe = vif_rfe.sort_values( by = 'VIF', ascending = False)
        
        if vif_rfe.iloc[0].VIF>=5:
            multicollinearity_detected=True
        
        maxtimes=maxtimes-1
    vif_func=vif_rfe    
    return vif_func
    
```
